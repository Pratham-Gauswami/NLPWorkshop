```python
df = pd.read_csv('/content/reddit_worldnews_start_to_2016-11-22.csv')
df
```

```python
df.shape #check out this part
```

```python
df.title[0]
gensim.utils.simple_preprocess("Scores killed in Pakistan clashes")
```

```python
news_title = df.title.apply(gensim.utils.simple_preprocess)
news_title
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2,
    workers = 4,
)
```

```python
#build voabulary function
model.build_vocab(news_title, progress_per=1000)
```

```python
model.epochs #an epoch refers to one complete pass through the entire training dataset. During an epoch, the model is exposed to all the training examples and updates its parameters based on the patterns it learns.
```

```python
model.wv.most_similar("coffee") #Calculate the similarity with the other wiords. (Can be used to train the model to get it familiar with a word with two meanings.)
```
